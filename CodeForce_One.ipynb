```python
def trapezoid_area(b1,b2,h):
  return (1/2) * (b1+b2) * h
height = 5
base_one = 5
base_two = 6
print(f"Area of trapezoid : {trapezoid_area(base_one,base_two,height)}")
```

```python
def parallelogram_area(h,b):
  return h * b
length = 5
height = 6
print(f'Parallelogram area is :{parallelogram_area(length,height)}')
```

```python
def cylinder(height,radious):
  print(f'Volums of cylinder is : {3.14 * (radious**2) * height}')
  print(f"Total surface area of cylinder is {2 * 3.14 * radious * (radious + height)}")
height = 4
radious = 6
cylinder(height,radious)
```

```python
def sphere(r):
  print(f"Surface Area of Sphere {4 * 3.14 * (r**2) }")
  print(f'Volume of Sphere: {(4/3) * 3.14 * (r**3)}')
r = .75
sphere(r)
```

```python
def arc(d,a):
  r = d/2
  print('Arc length in radian',(r*a))
  print('Arc length in degree',(a/360)*2*3.14*r)
d = 8
a = 45
arc(d,a)
```

```python
def sector_area(r,a):
  print('Sector Area',(a/360)*3.14*(r**2))
r= 4
a = 45
sector_area(r,a)
```
