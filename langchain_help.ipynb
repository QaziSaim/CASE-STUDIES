```python
"""
Problem
Create a program that calculates the area of different shapes (Circle, Rectangle, Triangle) using a common interface/class.
Approach
First i will create a base class
Then derived classes overrides the area
"""



class Shape:
    def area(self):
        pass

class Circle(Shape):
    PI = 3.14159
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return self.PI * self.radius * self.radius
        # shape.area = self.pie * (self.radius*self.radius)

class Rectangle(Shape):
    def __init__(self,w,h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h

class Trianlge(Shape):
    def __init__(self,b,h):
        self.b = b
        self.h = h
    def area(self):
        return 0.5 * self.b * self.h

if __name__=="__main__":
    shapes = [
    Circle(5),
    Rectangle(4, 6),
    Trianlge(3, 7)
]

    for shape in shapes:
        print("Area:", shape.area())

```
