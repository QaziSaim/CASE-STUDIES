```python
## Numba
from numba import jit
import math
```

```python
@jit
def hypot(x, y):
  x = abs(x);
  y = abs(y);
  t = min(x,y);
  x = max(x,y);
  t = t/x;
  return x * math.sqrt(1 + t*t)
```

```python
hypot(3.0,4.0)
```

```python
hypot.py_func(3.0, 4.0)
```

```python
%timeit hypot(3.0, 4.0)
```
